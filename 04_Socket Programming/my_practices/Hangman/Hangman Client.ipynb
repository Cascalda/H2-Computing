```python
import socket

game_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
print("Waiting to connect to server...")

game_socket.connect(('localhost', 12345))
print("Connected to Server")

def receive_msg(socket):
    rx_msg_str = ""
    while '\n' not in rx_msg_str:
        rx_msg_str += socket.recv(1024).decode()
    return rx_msg_str[:-1]
    

def display_menu():
    display = """
    SELECT AN OPTION:
    1. Guess a Letter
    2. Guess a Word
    3. Quit
    """
    
    return display

# Decoration
def string_generator(repeat_str, repeat_no):
    line = repeat_str * repeat_no
    return line

def question_marks(socket):
    received = receive_msg(socket).split(",")
    start = received[0] + ': '
    len_of_qn = int(received[1])
    qn_str = len_of_qn * "?"
    
    string = start + qn_str
    return string
    
def menu():
    print("Waiting for word...")
    ## Print Word in Prog
    guess_letter = ""
    
    receive_start = receive_msg(game_socket).split(",")
    print("_________________PROGRAME STARTED__________________")
    start_no = int(receive_start[1]) ## Receive the number
    start_lst = []
    
    for i in range(start_no):
        start_lst.append("?")
    start_qnmarks = ""
    for ele in start_lst:
        start_qnmarks += ele
    
    ## PRINT STARTING WORD
    print("START:", start_qnmarks)
    # print("start_lst", start_lst)
        
    done = False
    while not done:
        received = ""
        start_qnmarks = ""
        
        for ele in start_lst:
            start_qnmarks += ele
        
        ## PRINT CONTINUED WORD
        print("CONTINUE:", start_qnmarks)
        # print("start_lst", start_lst)

        line = string_generator("_", 50)
        print(line)
        
        display = display_menu()
        print(display)
            
        ## HANDLE USER INPUT
        user_option = input("Please input your option: ")
        print("You chose:", user_option)
            
        if user_option == "1":
            guess = input("Guess a letter: ")
            msg = 'GUESS,{}\n'.format(guess)
            game_socket.send(msg.encode())
            
            # [GUESS,0,2,...]
            received = receive_msg(game_socket)
            received_lst = received.split(",")
            num_lst = received_lst[1:]

            for pos in num_lst:
                start_lst[int(pos)] = guess
                
            # print("received_lst", received_lst)
            # print("num_lst", num_lst)
            # print("start_lst", start_lst)
            
        elif user_option == "2": 
            guess = input("Guess the word: ")
            msg = 'HWORD,{}\n'.format(guess)
            game_socket.send(msg.encode())
            
            received = receive_msg(game_socket)
        
        elif user_option == "3":
            msg = 'QUIT\n'
            game_socket.send(msg.encode())
            print("You Quit")
            done = True
        
        ## WIN OR LOSE
        if received == "WIN":
            print("YOU WIN :D")
            done = True

        elif received == "LOSE":
            print("YOU LOSE D:")
            done = True
    
    print("________________PROGRAME TERMINATED________________")
    game_socket.close()
        
menu()
```
